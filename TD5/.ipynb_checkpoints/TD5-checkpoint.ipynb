## 2. Analyse exploratoire


### Survie vs Classe des passagers

On souhaite vérifier s'il existe une relation entre la classe des passagers et leur taux de survie.

```julia
df = combine(groupby(data, :Pclass), :Survived => mean => Symbol("Taux de survie"))

# Permutation des colonnes pour mettre les classes dans le bon ordre pour le graphique
df = df[[1;3;2], :]  

set_default_plot_size(10cm, 8cm)
plot(df, x=:Pclass, y=Symbol("Taux de survie"), Geom.bar)
```

### Survie vs Genre des passagers

```julia
df = combine(groupby(data, :Sex), :Survived => mean => Symbol("Taux de survie"))

set_default_plot_size(10cm, 8cm)
plot(df, x=:Sex, y=Symbol("Taux de survie"), Geom.bar)
```

### Survie vs Age des passagers 

Exploration des données sur  l'âge des passagers.

```julia
(n,p) = size(data)

describe(data) # Age : 174 valeurs manquantes sur 872

# Entrée : df avec valeurs manquantes 
# Sortie : df avec valeurs manquantes remplacées par l'âge médian du df
function age_impute(df)
    Age_median = median(skipmissing(df.Age))
    replace!(df.Age, missing => Age_median)
end 

# Entrée : df sans catégorisation par tranche d'âge
# Sortie : df avec 4 catégorisation par tranche d'âge
# RQ : On crée une nouvelle colonne catégorielle (pour éviter le problème d'idempotence si on répète plusieurs fois la transformation sur la même colonne)
function age_cat(df)
    df[!, :AgeCat] = df[:, :Age]

    df.AgeCat[df.Age .< 2] .= 1 # Bébés
    df.AgeCat[2 .<= df.Age .< 28] .= 2 # Enfants 
    df.AgeCat[18 .<= df.Age] .= 3 # Adultes
    #df.AgeCat[65 .<= df.Age] .= 3 # Senior
    
    df[!, :AgeCat] = string.(df[:, :AgeCat]);
    
    replace!(df.AgeCat, 
        "1.0" => "Baby", 
        "2.0" => "Child",
        "3.0" => "Adult")
end 

age_impute(data)
age_cat(data)

"""
Idées en vrac pour l'âge :
1 - Catégoriser l'âge : Baby, Child, Adult, Senior
2 - Pour les valeurs manquantes : 
2.1 - Moyenne : Adulte 
2.2 - Affiner en fonction des Siblings ou d'autres var explicatives. 
"""
```

```julia
df = combine(groupby(data, :AgeCat), :Survived => mean => Symbol("Taux de survie"))

set_default_plot_size(10cm, 8cm)
plot(df, x=:AgeCat, y=Symbol("Taux de survie"), Scale.x_discrete(levels=["New", "Baby", "Child", "Adult"]), Geom.bar)
```

### Survive vs SibSp

```julia
df = combine(groupby(data, :SibSp), :Survived => mean => Symbol("Taux de survie"))

set_default_plot_size(10cm,8cm)
plot(df, x=:SibSp, y=Symbol("Taux de survie"), Coord.cartesian(xmin=-1,xmax=5), Geom.bar)
```

### Sibsp vs Age 

```julia

set_default_plot_size(10cm,8cm)
plot(data, x=:AgeCat, y=mean(:SibSp), Geom.point)
```

## 3. Modèle de régression logistique 

La classe des passagers est utilisée comme variable explicative de la probabilité de survie. La fonction de lien logit est utlisée.

```julia
M = glm(@formula(Survived ~ Pclass + Sex + AgeCat + SibSp), data,  Bernoulli(), LogitLink()) 
```

Note pour moi : 
- p-value petite (<0.05 par ex.) : On rejette H₀. On considère que le variable a un impact réel pour prédire la survie.
- p-value grande (>0.05 par ex.) : On rejette H₁. On ne peux pas exclure le fait que l'effet observé soit dû au pur hasard.

<a id="section_4"></a>
## 4. Prédiction de la survie des passagers de l'ensemble de test

### 4.1 Chargement des données de l'ensemble de test

```julia
# Charger le jeu de données de test
test = CSV.read("Test.csv", DataFrame);

# Conversion en String pour que GLM considère la classe comme une variable qualitative
test[!, :Pclass] = string.(test[:, :Pclass]);  

# Age pipeline
age_impute(test);
age_cat(test);
```

### 4.2 Estimation de la probabilité de survie de chacun des passagers de l'ensemble de test

```julia
θ̂ = predict(M, test) 
```

### 4.3 Prédiction de la survie des passagers de l'ensemble de test

```julia
# Prédiction de la survie pour chacun des passagers de l'échantillon de test
Ŷ = zeros(Int64, size(test,1))
Ŷ[θ̂.>.5] .= 1;
```

### 4.4 Préparation du fichier des préditions pour téléverser sur Kaggle

```julia
# Enregistrement dans un fichier CSV pour soumission sur Kaggle
# Première colonne : identifiant des passagers
# Deuxième colonne : survie = 1, décès = 0 (nombres entiers)
print("OK")
prediction = DataFrame(PassengerId = test[:,:PassengerId], Survived = Ŷ)
CSV.write("benchmark_predictions.csv", prediction)
```
