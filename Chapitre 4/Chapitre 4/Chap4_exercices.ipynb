```julia
using CSV, DataFrames, LinearAlgebra, Statistics
```

```julia
data = CSV.read("bodyfat.csv", DataFrame)
```

```julia
X̃ = Matrix(data[:,1:3])
mₓ = mean(X̃, dims=1)
sₓ = std(X̃, dims=1)

X = (X̃ .- mₓ)./ sₓ

ỹ = data[:,4]

m = mean(ỹ)
s = std(ỹ)

y = (ỹ .- m) ./ s
```

```julia
F = svd(X)
```

```julia
γ = F.S

p = cumsum(γ.^2) / sum(γ.^2)
```

```julia
T = X*F.V
```

```julia
η̂ = T\y
```

```julia
β̂ = F.V*η̂
```

```julia
x₀ = ([25.0 45.0 30.0] - mₓ) ./ sₓ 
```

```julia
y₀ = x₀*β̂*s .+ m
```
